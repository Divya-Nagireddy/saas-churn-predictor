```python
import pandas as pd
# 1. We import our charting libraries (The Python equivalent of Chart.js)
# 'sns' is the standard nickname for Seaborn. 'plt' is for Matplotlib (the canvas it draws on).
import seaborn as sns
import matplotlib.pyplot as plt
# 1. Import our AI tools from Scikit-Learn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
# 1. We import the "grading rubric" from Scikit-Learn
from sklearn.metrics import accuracy_score
# 1. Import our new algorithm from Scikit-Learn
from sklearn.ensemble import RandomForestClassifier
# 1. Import our "bundler" tool
import joblib


# 1. We've updated the URL to point to IBM's active repository for this dataset.
# Think of this as updating the endpoint in your JS fetch() call.
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"

# 2. We use Pandas to read the CSV file from the new URL.
data = pd.read_csv(url)

# 3. Let's see all the columns we have to work with.
# (The Pandas equivalent of Object.keys(data[0]))
print("--- Our Data Columns ---")
print(data.columns)

# 4. Let's look at the first 3 rows of a few key columns to see the data structure.
print("\n--- Snapshot of our Customers ---")
print(data[['customerID', 'tenure', 'MonthlyCharges', 'Churn']].head(3))


# TICKET TASK: The Pandas version of grouping and averaging.
# Read this as: "Take the data, group everyone by their 'Churn' status (Yes or No),
# then grab the 'tenure' and 'MonthlyCharges' columns, and calculate the mean (average)."
summary = data.groupby('Churn')[['tenure', 'MonthlyCharges']].mean()
print("---Average Stats: Stayed vs. Churned---")
print(summary)

# 2. We are going to draw a "Box Plot".
# This draws a box showing where the majority of the customers fall.
# x-axis = Did they churn? y-axis = How many months were they subscribed?
sns.boxplot(x='Churn', y='tenure', data=data)

# 3. Give the chart a clean title
plt.title('Customer Lifespan: Stayed vs Cancelled')

# 4. Render the chart to the screen
plt.show()

# 2. Prep the data (The JS Bridge: Mapping strings to booleans)
# AI models are basically giant calculators; they need numbers, not text.
# So, we convert "Yes" to 1 and "No" to 0.
data['Churn_Number'] = data['Churn'].map({'Yes':1, 'No': 0})

# 3. Define our Inputs (X) and our Output (y)
# X = The Clues we give the AI. y = The Answer we want it to learn.
x=data[['tenure', 'MonthlyCharges']]
y=data['Churn_Number']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

# 5. Create the blank "Brain"
# max_depth=3 means we only let the tree make 3 layers of decisions so it doesn't overthink.
model = DecisionTreeClassifier(max_depth=3)
# 6. Train the Brain
model.fit(x_train, y_train)
print("🧠 The AI has been successfully trained!")

# 2. The Final Exam!
# We hand the AI the test questions (X_test) and tell it to make its guesses.
predictions = model.predict(x_test)


# 3. The Grading!
# We compare the AI's guesses against the actual real-world answers (y_test).
score = accuracy_score(y_test, predictions)

# 4. Print the final grade to the screen.
# We multiply by 100 to make it a readable percentage.
print(f"🎯 AI Final Exam Score: {score * 100:.2f}%")

# 1. We grab a bunch of extra columns this time!
# Some are numbers (tenure, MonthlyCharges), some are text (Contract, InternetService).
features = data[['tenure', 'MonthlyCharges', 'Contract', 'InternetService', 'TechSupport']]

# 2. THE TRANSLATOR (One-Hot Encoding)
# pd.get_dummies() looks at all the text columns and automatically turns them into 1s and 0s (True/False).
x_upgraded = pd.get_dummies(features)

# Let's look at the new columns Pandas generated for us!
print("--- Upgraded AI Clues (Notice the new 'Checkbox' columns) ---")
print(x_upgraded.head(3))

# 1. We grab our "Answers" again just to be safe (1 = Churned, 0 = Stayed)
y = data['Churn_Number']

# 2. The New Split: We use our 'x_upgraded' data this time!
x_train_up, x_test_up, y_train_up, y_test_up = train_test_split(x_upgraded, y, test_size=0.2, random_state=42)

# 3. Build a slightly bigger brain!
# Since we have more data, we increase 'max_depth' to 5.
# This lets the AI create a slightly deeper chain of if/else statements.
model_upgraded = DecisionTreeClassifier(max_depth=5, random_state=42)

# 4. TEACH THE AI (The Practice Test)
model_upgraded.fit(x_train_up, y_train_up)

# 4. TEACH THE AI (The Practice Test)
model_upgraded.fit(x_train_up, y_train_up)

# 5. GRADE THE AI (The Final Exam)
predictions_up = model_upgraded.predict(x_test_up)

score_up = accuracy_score(y_test_up, predictions_up)

print(f"🚀 UPGRADED AI Final Exam Score: {score_up * 100:.2f}%")

# 2. Build the Board of Directors
# 'n_estimators=100' simply means "Build 100 different Decision Trees"
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 3. TEACH THE AI (The Practice Test)
# We are using the exact same upgraded data from the last step.
# It might take 1 or 2 seconds longer this time because it's training 100 brains!
forest_model.fit(x_train_up, y_train_up)

# 4. GRADE THE AI (The Final Exam)
forest_predictions = forest_model.predict(x_test_up)

forest_score = accuracy_score(y_test_up, forest_predictions)

print(f"🌲 Random Forest Final Exam Score: {forest_score * 100:.2f}%")

# 2. We take our best trained brain (model_upgraded) and save it to a physical file.
# We are naming the file 'saas_saver_brain.joblib'
joblib.dump(model_upgraded, 'saas_saver_brain.joblib')

print("📦 AI Brain successfully bundled and saved!")

# 3. (Optional) In Google Colab, you can look at the file folder icon on the far left side
# of your screen. You will see your new 'saas_saver_brain.joblib' file sitting there!
```
